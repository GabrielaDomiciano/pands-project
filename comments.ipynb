```python
# Exploratory Analysis: Iris Data_set
# Import required packages:
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Let's take a look at the Dataset.
iris_df = pd.read_csv('iris-data.csv')
print(iris_df)
```

```python
# Empty columns were deleted, deleting these columns helps in viewing the DataFrame.
iris_df = iris_df.dropna()
print(iris_df)
```

```python
# Calculates descriptive statistics, and saves it to a CSV file called 'summary.csv'
iris_df.describe().T.to_csv('summary.csv')
```

```python
# Print the first 10 lines of the DataFrame
print(iris_df.head(10))
```

```python
# Print the shape, number of rows and columns in the DataFrame.
print(iris_df.shape)
```

```python
# Checks for missing values ​​in any column of DataFrame 'iris_df'
missing_values = iris_df.isna().any()
print(missing_values)
```

```python
# Inspect types.
type = iris_df.dtypes
print(type)
```

```python
# Lets see the petal length.
petal_length = iris_df['petal_length']
print(petal_length)
```

```python
# Lets see the petal width.
petal_width = iris_df['petal_width']
print(petal_width)
```

```python
## Number the flower per species.
count = iris_df['species'].value_counts()
print(count)
iris_df['species'].value_counts().plot(kind='bar')
plt.title('Number of Species')
plt.legend(title='Quantity per Species', labels=count)
plt.savefig('Quantity per Species')
plt.show()
```

```python
# The petal_length per species.
petal_length = iris_df[['species', 'petal_length']]
# Show
print(petal_length)
```

```python
# The sepal_length per species.
sepal_length = iris_df[['species', 'sepal_length']]
# Show
print(sepal_length)
```

***

```python
# https://eminebozkus.medium.com/exploring-the-iris-flower-dataset-4e000bcc266c
# Checking for the unique values
iris_df.nunique()
```

```python
# Visualize the whole dataset
# https://eminebozkus.medium.com/exploring-the-iris-flower-dataset-4e000bcc266c
# grid of scatter plots showing the relationships between different pairs of features in the Iris dataset, with points colored according to the species of the Iris flower.
# This can be helpful for visualizing the relationships between variables and for exploring the dataset.
# 

sns.pairplot(iris_df,hue="species")
plt.show()
```

There is a connection between the length and width of the petal. The Setosa species has a low petal length and width, the Versicolor species has both a medium petal length and width. The Virginica species has both tall petal length and width. In terms of sepal dimensions, the Setosa species has a high sepal width and a low sepal length, the Versicolor species has medium sepal values, and the Virginica species has a small sepal width but a large sepal length.

```python
# https://gist.github.com/mwaskom/de44147ed2974457ad6372750bbe5751

# The code generates four plots, one for each feature in the Iris dataset: sepal length, sepal width, petal length, 
# and petal width. Each one shows the distribution of the respective characteristic in the data set.

# Plotting the distplot 
sns.histplot(iris_df['sepal_length']) #[1]
plt.show()

sns.histplot(iris_df['sepal_width']) #[2]
plt.show()

sns.histplot(iris_df['petal_length'])#[3]
plt.show()

sns.histplot(iris_df['petal_width']) #[4]
plt.show()
```

[1] - Sepal Length: Most of the samples have a sepal length between approximately 4.5 and 7.0. There appears to be a normal distribution, with a peak around 5.0 to 6.0.

[2] - Sepal Width: Most samples have a sepal width between approximately 2.5 and 4.0. The distribution appears to be slightly less symmetrical than sepal length, with some more frequent values ​​around 3.0.

[3] - Petal Length: It shows a more distribution. Most samples have a petal length between 1.0 and 6.0. There appear to be two distinct groups of samples, with one group around 1.0 to 2.0 and another larger group around 4.5 to 6.0.

[4] - Petal Width: We see a similar distribution to petal length. Most samples have a petal width between approximately 0.1 and 2.5. There appear to be two distinct groups of samples, with a larger group around 0.1 to 0.6 and another smaller group around 1.5 to 2.5.

***

```python
# The plots below show the relationships between the width and length of petals and sepals of species.

# Sets the figure size and creates a scatter plot showing the relationship between petal width and length,
# with different iris species represented by different colors.
plt.figure(figsize=(8,6))
sns.scatterplot(data = iris_df, x ='petal_length', y ='petal_width', hue ='species')
plt.title('Petals Width vs Length')
plt.xlabel('Width')
plt.ylabel('Length')
plt.savefig('Petals Width vs Length')
plt.show()
```

```python
# This code creates a scatterplot using seaborn to visualize the relationship 
# between sepal length and petal width of flowers, with different colors for each species.
plt.figure(figsize=(8,6))
sns.scatterplot(data = iris_df, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.title('Sepals Width vs Length')
plt.xlabel('Width')
plt.ylabel('Length')
plt.savefig('Sepals Width vs Length')
plt.show()
```
