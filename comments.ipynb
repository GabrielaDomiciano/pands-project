## Load Data
*** 

The Iris Data consists of four characteristics: sepal length, sepal width, petal length, and petal width. These resources provide information about the sizes of Iris flowers.
The dataset comprises three species: Setosa, Versicolor and Virginica. Each species is represented by 50 samples, with a total of 150.

```python
# Let's take a look at the Dataset.
iris_df = pd.read_csv('iris-data.csv')
print(iris_df)
```

There is a connection between the length and width of the petal. The Setosa species has a low petal length and width, the Versicolor species has both a medium petal length and width. The Virginica species has both tall petal length and width. In terms of sepal dimensions, the Setosa species has a high sepal width and a low sepal length, the Versicolor species has medium sepal values, and the Virginica species has a small sepal width but a large sepal length.

```python
# https://gist.github.com/mwaskom/de44147ed2974457ad6372750bbe5751

# The code generates four plots, one for each feature in the Iris dataset: sepal length, sepal width, petal length, 
# and petal width. Each one shows the distribution of the respective characteristic in the data set.

# Creates a histogram of sepal lengths from the Iris dataset, with a KDE curve overlaid on top.
# "KDE" will include a density estimation curve at the top of the histogram bars. 
# When set to True, a smooth curve representing the distribution of the data is overlaid.

# Plotting the distplot 
sns.histplot(iris_df['sepal_length'], color='pink') #[1]
sns.histplot(iris_df["sepal_length"], kde=True, color='blue')
plt.show()

sns.histplot(iris_df['sepal_width'], color='pink') #[2]
sns.histplot(iris_df["sepal_width"], kde=True, color='blue')
plt.show()

sns.histplot(iris_df['petal_length'], color='pink')#[3]
sns.histplot(iris_df["petal_length"], kde=True, color='blue')
plt.show()

sns.histplot(iris_df['petal_width'], color='pink') #[4]
sns.histplot(iris_df["petal_width"], kde=True, color='blue')
plt.show()
```

[1] - Sepal Length: Most of the samples have a sepal length between approximately 4.5 and 7.0. There appears to be a normal distribution, with a peak around 5.0 to 6.0.

[2] - Sepal Width: Most samples have a sepal width between approximately 2.5 and 4.0. The distribution appears to be slightly less symmetrical than sepal length, with some more frequent values ​​around 3.0.

[3] - Petal Length: It shows a more distribution. Most samples have a petal length between 1.0 and 6.0. There appear to be two distinct groups of samples, with one group around 1.0 to 2.0 and another larger group around 4.5 to 6.0.

[4] - Petal Width: We see a similar distribution to petal length. Most samples have a petal width between approximately 0.1 and 2.5. There appear to be two distinct groups of samples, with a larger group around 0.1 to 0.6 and another smaller group around 1.5 to 2.5.

```python
# https://gist.github.com/mwaskom/de44147ed2974457ad6372750bbe5751
# The plots below show the relationships between the width and length of petals and sepals of species.

# Sets the figure size and creates a scatter plot showing the relationship between petal width and length,
# with different iris species represented by different colors.
plt.figure(figsize=(8,6))
sns.scatterplot(data = iris_df, x ='petal_length', y ='petal_width', hue ='species')
plt.title('Petals Width vs Length')
plt.xlabel('Width')
plt.ylabel('Length')
plt.savefig('Petals Width vs Length')
plt.show()
```

The scatter plot compares the length and width of the petals of the iris species. Each dot represents a flower sample.
We observed a correlation between the length and width of the petal, which means that as the length of the petal increases, the width tends to increase as well, remaining the same in all species.
The dots of the “setosa” species form a cluster in the lower left corner and have shorter, narrower petals compared to the other species.
The species “versicolor” and “virginica” have the most similar petal characteristics compared to “setosa”.
The scatterplot demonstrates how petal length and width can be used to differentiate iris species, with the "setosa" having unique petal characteristics relative to other species. (the dataset contains 150 samples some samples may be overlapping)



```python
# This code creates a scatterplot using seaborn to visualize the relationship 
# between sepal length and petal width of flowers, with different colors for each species.
plt.figure(figsize=(8,6))
sns.scatterplot(data = iris_df, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.title('Sepals Width vs Length')
plt.xlabel('Width')
plt.ylabel('Length')
plt.savefig('Sepals Width vs Length')
plt.show()
```

The scatterplot compares the length and width of the Iris sepals.
The color of the dots allows us to see how the characteristics of the sepals vary between species.
The horizontal line separating the species is less clear than on the petal chart, which may suggest that the sepals may not be as different between species as the petals.
While there is some overlap between species, we can still observe general patterns. For example, flowers of the "setosa" species generally have shorter, wider sepals, while the "versicolor" and "virginica" species have longer, narrower sepals.(the dataset contains 150 samples some samples may be overlapping)
